#Library dan Modul

Kode yang kami gunakan mengimpor tiga library kunci: pandas untuk manipulasi data, requests untuk mengirimkan HTTP requests, dan BeautifulSoup untuk scraping data dari HTML dan XML. Dengan kombinasi ini, kami dapat mengambil data dari halaman web menggunakan requests, kemudian memproses dan menganalisis data menggunakan pandas, serta melakukan scraping data dari HTML dan XML menggunakan BeautifulSoup.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

#Scraping

Situs web "https://www.tempo.co" adalah portal berita yang terkenal di Indonesia yang menyediakan liputan luas mengenai berbagai topik, termasuk berita politik, ekonomi, sosial, budaya, olahraga, dan banyak lagi. Tempo.co adalah bagian dari Tempo Media Group, sebuah perusahaan media terkemuka di Indonesia. Situs ini menawarkan artikel-artikel berita terkini, laporan investigasi mendalam, serta opini dari penulis dan pakar di berbagai bidang.Halaman ini menampilkan daftar artikel yang paling populer atau paling banyak dilihat oleh pengunjung situs dalam periode waktu tertentu.

```python
# Mendapatkan page HTML melalui request
page = requests.get('https://www.tempo.co/terpopuler')

# Parsing konten halaman menggunakan beautifulsoup
page_parsed = BeautifulSoup(page.text, 'html.parser')

page_parsed
```

```python
judul = page_parsed.select("div article h2 a")
judul
```

```python
data=[]
for juduls in judul:
  data.append(juduls.get_text())

df = pd.DataFrame(data, columns=['Judul_Berita'])
print(df)
```

```python
deskripsi = page_parsed.select("article p")
deskripsi
```

```python
data2=[]
for desc in deskripsi:
  data2.append(desc.get_text())

df2 = pd.DataFrame(data2, columns=['Deskripsi'])
print(df2)
```

```python
tanggal= page_parsed.findAll("h4", {"class": "date"})
tanggal
```

```python
data3=[]
for dates in tanggal:
  data3.append(dates.get_text())

df3 = pd.DataFrame(data3, columns=['Tanggal'])
df3[:6]
```

```python
hasil_data = pd.concat([df,df2,df3], axis=1)
hasil_data
```

#Import To Excel

```python
import pandas as pd
# Mengonversi data menjadi DataFrame Pandas
df = pd.DataFrame(hasil_data)

# Menyimpan DataFrame ke dalam file CSV
df.to_csv('hasil_scraping.csv', index=False)

print("Data telah disimpan ke dalam file CSV.")
```
