```python
import requests
```

```python
url = "https://www.olx.co.id/api/relevance/v4/search?category=208&facet_limit=100&location=4000024&location_facet_limit=20&platform=web-desktop&query=handphone&relaxedFilters=true&spellcheck=true&user=18e980c4fd9x6843dcdd"
payload = {}
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"}

response = requests.request("GET", url, headers=headers, data=payload)
print(response.text)
```

```python
# GANTI TIPE KE JSON
import json

json_data = json.loads(response.text)['data']
```

```python
#Tabulasi
import pandas as pd
df = pd.json_normalize(json_data)
df
```
